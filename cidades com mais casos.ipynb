```python
# Importação de bibliotecas e módulos
import pandas as pd
import numpy as npb
```

```python
# Abrindo como str a coluna "city_ibge_code"
df = pd.read_csv("caso_full.csv", dtype = { "city_ibge_code" : str}, parse_dates=['date'])
```

```python
# Vendo meu df
df.head()
```

```python
# Filtrando apenas os últimos dias. == true quer dizer que é o último
df = df[df.is_last == True]
```

```python
# Tirando as colunas que não servem para a análise
df = df.drop(["estimated_population_2019", "is_repeated", "is_last","last_available_confirmed_per_100k_inhabitants", "last_available_date", "last_available_death_rate", "place_type", "new_confirmed", "new_deaths", "order_for_place"], axis=1)
df.head()
```

```python
# Modificando os nomes das colunas
df.columns = ['municipio', 'código_ibge', 'data','Semana_epidemiologica', 'populacao', 'casos', 'obitos', 'uf']
df.head()
```

```python
# Criando coluna %casos_pop, que mostrará qual a porcentagem daquela população já foi infectada pelo coronavírus
df["%casos"] = (df.casos/df.populacao)*100
```

```python
# Deixando as maiores porcentagem de maior para menor 
casos = df.sort_values('%casos', ascending=False)
casos.head(10)
```
